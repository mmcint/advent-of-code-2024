# Part 1

```python
import pandas as pd
import numpy as np
```

```python
data = pd.read_csv('day1_data.txt',sep="\s+",header=None, names=["Column1", "Column2"])
```

```python
out_df = pd.concat([data['Column1'].sort_values(ascending=True).reset_index(drop=True), data['Column2'].sort_values(ascending=True).reset_index(drop=True)], axis=1)
```

```python
out_df
```

```python
out_df['diff'] = np.abs(out_df['Column1'] - out_df['Column2'])
```

```python
print('Answer: {}'.format(np.sum(out_df['diff'])))
```

# Part 2

```python
score_dict = {}
count = 0
for i in list(data['Column1']):
  count = 0
  for j in list(data['Column2']):
    if i == j:
      count += 1
    else:
      pass
    score_dict[i] = count
```

```python
sim_score = 0
for k, v in score_dict.items():
  sim_score += (k * v)
print('Answer: {}'.format(sim_score))
```
